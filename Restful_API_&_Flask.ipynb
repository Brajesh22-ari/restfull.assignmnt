```python
1.What is a RESTful API?
ANS. A RESTful API (Representational State Transfer API) is a way for computer systems to communicate over the internet using standard HTTP methods like:

GET – to retrieve data

POST – to create data

PUT – to update data

DELETE – to remove data
Key Features:
Stateless: Each request from the client to the server must contain all the information needed; the server doesn’t store session data.

Client-Server Separation: The frontend (client) and backend (server) work independently.

Uniform Interface: Resources (like users, posts, products) are accessed via URLs, such as:

GET /users – list users

GET /users/1 – get user with ID 1

POST /users – create a new user

Uses standard formats: Most RESTful APIs return data in JSON or XML formats.

Cacheable: Responses can be stored to reduce server load and improve speed.
Example:
If you’re building an app for managing books:

GET /books → returns a list of books

POST /books → adds a new book

GET /books/5 → returns book with ID 5

PUT /books/5 → updates book with ID 5

DELETE /books/5 → deletes book with ID 5
```

```python
14.How do you create a RESTful API endpoint using Flask?
ANS.A RESTful API endpoint in Flask is a route that responds to HTTP methods like GET, POST, PUT, and DELETE. You can build these using basic Flask functions and the @app.route() decorator.

Step-by-Step Example: Simple RESTful API
. Install Flask
pip install flask
2. Basic Flask REST API Code
from flask import Flask, jsonify, request

app = Flask(__name__)

# Mock database
users = [
    {"id": 1, "name": "Alice"},
    {"id": 2, "name": "Bob"}
]

# GET all users
@app.route('/users', methods=['GET'])
def get_users():
    return jsonify(users)

# GET user by ID
@app.route('/users/<int:user_id>', methods=['GET'])
def get_user(user_id):
    user = next((u for u in users if u['id'] == user_id), None)
    return jsonify(user) if user else ("User not found", 404)

# POST (create) a new user
@app.route('/users', methods=['POST'])
def add_user():
    data = request.get_json()
    new_user = {
        "id": users[-1]['id'] + 1 if users else 1,
        "name": data['name']
    }
    users.append(new_user)
    return jsonify(new_user), 201

# PUT (update) a user
@app.route('/users/<int:user_id>', methods=['PUT'])
def update_user(user_id):
    data = request.get_json()
    user = next((u for u in users if u['id'] == user_id), None)
    if user:
        user['name'] = data['name']
        return jsonify(user)
    return ("User not found", 404)

# DELETE a user
@app.route('/users/<int:user_id>', methods=['DELETE'])
def delete_user(user_id):
    global users
    users = [u for u in users if u['id'] != user_id]
    return ("", 204)

# Run the app
if __name__ == '__main__':
    app.run(debug=True)
```

```python
5.How can you generate URLs for routes in Flask using url_for?
ANS. use url_for()?
Automatically updates if routes change

Prevents broken links

Cleaner and safer than hardcoding URLs
Basic Syntax
url_for('function_name', **params)
Example 1: Basic Route
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/about')
def about():
    return "About Page"

@app.route('/')
def home():
    return f"Go to About: {url_for('about')}"
Example 2: Route with Parameters
@app.route('/user/<username>')
def profile(username):
    return f"Profile: {username}"

@app.route('/')
def index():
    return url_for('profile', username='brajesh')
| Use Case        | Example                                  |
| --------------- | ---------------------------------------- |
| Route URL       | `url_for('home')`                        |
| With parameters | `url_for('profile', username='brajesh')` |
| Static file     | `url_for('static', filename='main.js')`  |
```

```python
6.How do you handle forms in Flask?
ANS.Flask allows you to handle HTML forms easily using the request object to get form data from POST requests.

. Basic Form Flow in Flask
project/
├── app.py
└── templates/
    ├── form.html
    └── result.html
2. HTML Form Template (templates/form.html)
<!DOCTYPE html>
<html>
<head>
    <title>Form Example</title>
</head>
<body>
    <h1>Enter Your Name</h1>
    <form method="POST" action="/submit">
        <input type="text" name="username" placeholder="Your name">
        <input type="submit" value="Submit">
    </form>
</body>
</html>
 3. Flask App to Handle Form (app.py)
 from flask import Flask, request, render_template

app = Flask(__name__)

@app.route('/')
def index():
    return render_template('form.html')

@app.route('/submit', methods=['POST'])
def submit():
    username = request.form['username']
    return render_template('result.html', name=username)
 4. Result Template (templates/result.html)
 <!DOCTYPE html>
<html>
<head>
    <title>Result</title>
</head>
<body>
    <h2>Hello, {{ name }}!</h2>
</body>
</html>
5. How It Works:
User visits / and sees the form.

On submit, a POST request is sent to /submit.

Flask grabs the input using request.form['username'].

A new page shows the result using a rendered template.
```

```python
7.How can you validate form data in Flask?
ANS.two main ways to validate form data in Flask:
 Manual Validation (Basic, No Extra Packages)
 from flask import Flask, request, render_template

app = Flask(__name__)

@app.route('/', methods=['GET', 'POST'])
def register():
    error = None
    if request.method == 'POST':
        username = request.form.get('username')
        if not username:
            error = "Username is required"
        elif len(username) < 3:
            error = "Username must be at least 3 characters"
        else:
            return f"Welcome, {username}!"

    return render_template('form.html', error=error)
templates/form.html
<form method="POST">
    <input type="text" name="username" placeholder="Username">
    <input type="submit" value="Submit">
    {% if error %}
        <p style="color:red;">{{ error }}</p>
    {% endif %}
</form>
Using Flask-WTF (Recommended for Large Projects)
pip install flask-wtf
Example:
from flask import Flask, render_template, redirect
from flask_wtf import FlaskForm
from wtforms import StringField, SubmitField
from wtforms.validators import DataRequired, Length

app = Flask(__name__)
app.secret_key = 'your-secret-key'  # Needed for CSRF protection

class NameForm(FlaskForm):
    name = StringField('Name', validators=[DataRequired(), Length(min=3)])
    submit = SubmitField('Submit')

@app.route('/', methods=['GET', 'POST'])
def home():
    form = NameForm()
    if form.validate_on_submit():
        return f"Hello, {form.name.data}!"
    return render_template('form_wtf.html', form=form)
templates/form_wtf.html
<form method="POST">
    {{ form.hidden_tag() }}
    {{ form.name.label }} {{ form.name() }}
    {% for error in form.name.errors %}
        <span style="color:red;">{{ error }}</span>
    {% endfor %}
    <br>
    {{ form.submit() }}
</form>

```

```python
8. How do you manage sessions in Flask?
ANS.Flask provides a built-in session object to store and manage user-specific data across requests (like login status, preferences, etc.).

A session stores data per user.

Flask uses signed cookies (via a secret key) to store session data on the client side.

You can access session data like a Python dictionary.

Setup: Secret Key Required
from flask import Flask, session

app = Flask(__name__)
app.secret_key = 'your-secret-key'  # Required for session security
Setting and Accessing Session Data
from flask import Flask, session, redirect, url_for, request

app = Flask(__name__)
app.secret_key = 'your-secret-key'

@app.route('/')
def index():
    username = session.get('username')
    if username:
        return f'Logged in as {username}'
    return 'You are not logged in'

@app.route('/login', methods=['GET', 'POST'])
def login():
    if request.method == 'POST':
        session['username'] = request.form['username']
        return redirect(url_for('index'))
    return '''
        <form method="post">
            <input type="text" name="username">
            <input type="submit" value="Login">
        </form>
    '''

@app.route('/logout')
def logout():
    session.pop('username', None)
    return redirect(url_for('index'))
| Action          | Code Example                  |
| --------------- | ----------------------------- |
| Set session key | `session['user'] = 'brajesh'` |
| Get session key | `session.get('user')`         |
| Remove key      | `session.pop('user', None)`   |
| Clear all data  | `session.clear()`             |
```

```python
11. How do you structure a Flask app using Blueprints?
ANS.Flask Blueprints allow you to organize your application into modular components, making it easier to scale, maintain, and reuse code.

A Blueprint is like a mini app. It contains its own:

Routes

Templates

Static files

Forms, models, etc.

You register the blueprint in your main Flask application.

myapp/
│
├── app.py                # Main entry point
├── home/
│   ├── __init__.py       # Blueprint setup
│   └── routes.py         # Routes for "home"
├── auth/
│   ├── __init__.py       # Blueprint setup
│   └── routes.py         # Routes for "auth"
├── templates/
│   └── ...
└── static/
    └── ...
Step-by-Step Setup
from flask import Blueprint, render_template

home_bp = Blueprint('home', __name__)

@home_bp.route('/')
def homepage():
    return render_template('home.html')
2. home/__init__.py
from .routes import home_bp
3. auth/routes.py
from flask import Blueprint, render_template

auth_bp = Blueprint('auth', __name__, url_prefix='/auth')

@auth_bp.route('/login')
def login():
    return render_template('login.html')
4. auth/__init__.py
from .routes import auth_bp
5. app.py (Main App)
from flask import Flask
from home import home_bp
from auth import auth_bp

app = Flask(__name__)

# Register Blueprints
app.register_blueprint(home_bp)
app.register_blueprint(auth_bp)

if __name__ == '__main__':
    app.run(debug=True)
| Feature               | Advantage                                |
| --------------------- | ---------------------------------------- |
| Modular Codebase      | Organize by feature or section           |
| Reusability           | Blueprints can be reused across projects |
| Scalable Architecture | Ideal for large applications             |
| Cleaner Imports       | Avoids circular import issues            |

```
